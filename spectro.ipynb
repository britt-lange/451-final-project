```python
import numpy as np
from matplotlib import pyplot as plt
import librosa
```

```python
#load example audio files
y_trumpet, sr_trumpet = librosa.load(librosa.ex('trumpet'))
y_hump, sr_hump = librosa.load(librosa.ex('humpback'))
```

```python
#function to plot spectrograms
def plot(y, sr):
    plt.figure(figsize=(10, 4))  # Adjust figsize as needed
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
    img = librosa.display.specshow(D, y_axis='linear', x_axis='time', sr=sr)
    plt.colorbar(img, format="%+2.f dB")
    plt.title('Linear-frequency power spectrogram')
    plt.xlabel('Time')
    plt.ylabel('Frequency (Hz)')
    plt.show()
    
#plot examples
plot(y_trumpet,sr_trumpet)
plot(y_hump, sr_hump)
```
